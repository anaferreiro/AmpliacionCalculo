Calcular 
$$
\lim_{(x,y)\to (0,0)}\frac{x^2 y^2 }{x^4 + y^4}
$$

```python
#Limites direccionales
import sympy as sp
x,y=sp.symbols('x,y')
k=sp.symbols('k')
f=(x**2 * y**2)/(x**4+y**4)
g=f.subs(y,k*x)
print(g)
#calculamos el limite direccional, considerando y=kx
l=sp.limit(g,x,0)
print(l)
#hay dependencia de la direcciÃ³n, por tanto, el lÃ­mite no existe
```

Calcular el lÃ­mite
$$
\lim_{(x,y)\to (0,0)}\frac{xy^3}{x^2 + y ^9}
$$

```python
#No hacer este ejercicio, porque tiene una peculiaridad
x,y=sp.symbols('x,y')
k=sp.symbols('k')
f=(x*y**3)/(x**2+y**9)
#limites en rectas
g=f.subs(y,k*x)
l1=sp.limit(g,x,0)
print(l1) #Puesto que no depende de l1, no concluye nada
#limite sobre parabolas
g=f.subs(y,k*x**2)
l2=sp.limit(g,x,0)
print(l2)#Puesto que no depende de l1, no concluye nada
g=f.subs(x,k*y**2)
l2=sp.limit(g,y,0)
print(l2)#Puesto que no depende de l1, no concluye nada
#cambio a coordenadas polares
ro,theta=sp.symbols('ro,theta')
g=f.subs([(x,ro*sp.cos(theta)),(y,ro*sp.sin(theta))])
print(g)
l3=sp.limit(g,ro,0)
print(l3)  
#Aunque el lÃ­mite es independiente del valor de theta, se observa que el 
#denominador de f(ro, theta) podrÃ­a anularse para algunos valores.. Basta 
#seguir la trayectoria x=ay**3 y probar que el lÃ­mite depende del parÃ¡metro
g=f.subs(x,k*y**3)
l4=sp.limit(g,y,0)
print(l4)
```
