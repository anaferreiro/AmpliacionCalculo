Ejr: Dada la curva
$$
c(t)=(3\cos^3(t), 3 \sin^3(t)) \quad t\in [0,pi/2]
$$

a) Calcular la longitud de la curva

b) Representar la curva

c) CAlcular una aproximacion de la longitud de la curva. Calcular el error cometido

```python
import sympy as sp
t=sp.symbols('t')
c=[3*sp.cos(t)**3,3*sp.sin(t)**3]
dc=[sp.diff(c[k],t) for k in [0,1]]
print(dc)
f=sp.sqrt(dc[0]**2+dc[1]**2 )
print(f)
f=sp.simplify(f)
print("f= ", f)
L=sp.integrate(f,(t,0,sp.pi/2), manual=True) #No es capaz de hacerla
L=sp.integrate(9*sp.sin(t)*sp.cos(t),(t,0,sp.pi/2), manual=True)
print("Longitud= ", L)
```

```python
import numpy as np
import matplotlib.pyplot as plt


ax = plt.figure().add_subplot(projection='3d')

# Prepare arrays x, y, z
t = np.linspace(0, 2*np.pi, 100)
 
x = 3 * np.cos(t)**3
y = 3 * np.sin(t)**3
z =0.*x

ax.plot(x, y, z, lw= 3,label='parametric curve')
ax.legend()

plt.show()


```

```python
import mpl_toolkits.mplot3d.art3d as art3d
ax = plt.figure().add_subplot(projection='3d')


n=3
a=0; b=np.pi/2
#dt=(b-a)/n
pt=np.linspace(a,b,n+1)
#print(pt, ", ", [0,np.pi/6,np.pi/3,np.pi/2])
def c(ti):
    return np.array([3.*np.cos(ti)**3,3.*np.sin(ti)**3])

curve=[]
for ti in pt:
    curve.append(c(ti))
#print(curve)

ax.plot(x, y, z, lw= 3,label='parametric curve')

for i in range(0,len(curve)-1):
    #print(i)
    xi,yi=curve[i]
    xj,yj=curve[i+1]
    line=art3d.Line3D( xs=[xi,xj], ys=[yi,yj], zs=[0,0] ,marker='o',color='red')
    ax.add_line(line)
    
plt.show()
```

```python
from numpy import linalg as LA



L_aprox=0.
for i in range(0, n):
    segment=curve[i+1]-curve[i]
    norma=LA.norm(segment)
    L_aprox+=norma
print("Longitud aproximada= ", L_aprox)
print("error= ", np.abs(L.evalf()-L_aprox))
```

Ejr: Dada la curva
$$
c(t)=(6\cos(t), 6 \sin(t),t^2) \quad t\in [0,pi/2]
$$

a) Calcular la longitud de la curva

b) Representar la curva

c) CAlcular una aproximacion de la longitud de la curva. Calcular el error cometido

```python
import numpy as np
import matplotlib.pyplot as plt


ax = plt.figure().add_subplot(projection='3d')

# Prepare arrays x, y, z
t = np.linspace(0, 6*np.pi, 100)
 
x = 6 * np.cos(t)
y = 6 * np.sin(t)
z =t**2

ax.plot(x, y, z, lw= 3)


plt.show()
```

```python
import sympy as sp
t=sp.symbols('t')
c=[6*sp.cos(t),6*sp.sin(t),t**2]
dc=[sp.diff(c[k],t) for k in [0,1,2]]
print(dc)
f=sp.sqrt(dc[0]**2+dc[1]**2 +dc[2]**2)
print(f)
f=sp.simplify(f)
print("f= ", f)
L=sp.integrate(f,(t,0,6*sp.pi)) 
print("Longitud= ", L.evalf())
```
