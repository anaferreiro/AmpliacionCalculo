Consideremos la funci√≥n:
$$
f(x,y)=\sqrt{64 -x^2 - y^2}
$$

```python
#%matplotlib qt
%matplotlib inline

import sympy as sp
x,y=sp.symbols('x,y')
f=sp.sqrt(64 - x**2 - y**2)
from sympy.plotting import plot3d
#Representacion grafica
fig=plot3d(f,(x,-5,5),(y,-5,5))
```

```python
#curvas de nivel con Sympy
from sympy.plotting.plot import ContourSeries, Plot
fig1=Plot(ContourSeries(f,(x,-5,5),(y,-5,5)))
fig1.show()
```

```python
#Curvas de nivel con Numpy
import numpy as np

import matplotlib.pyplot as plt
g=sp.lambdify((x,y),f,'numpy')
a=np.linspace(-5,5,100)
b=np.linspace(-5,5,100)
X,Y=np.meshgrid(a,b)
c=plt.contour(X,Y,g(X,Y))
```

```python
#Curvas de nivel con etiquetas con Numpy

fig,ax=plt.subplots()
cs=ax.contour(X,Y,g(X,Y))
ax.clabel(cs,inline=1,fontsize=10)
ax.set_title("con etiquetas")
```

Ej2: Consideramos la funci√≥n 
$$
f(x,y)=y^2 -x^2
$$

```python
x,y=sp.symbols('x,y')
f=y**2-x**2
from sympy.plotting import plot3d
#Representacion grafica
fig=plot3d(f,(x,-10,10),(y,-10,10))
```

```python
#Curvas de nivel con Numpy
import numpy as np

import matplotlib.pyplot as plt
g=sp.lambdify((x,y),f,'numpy')
a=np.linspace(-25,25,500)
b=np.linspace(-25,25,500)
X,Y=np.meshgrid(a,b)
levels = np.arange(-55., 55, 20)
c=plt.contour(X,Y,g(X,Y),levels=levels)



fig,ax=plt.subplots()
cs=ax.contour(X,Y,g(X,Y))
ax.clabel(cs,inline=1,fontsize=10)
ax.set_title("con etiquetas")
```

Ej3:  Un ejemplo de funci√≥n de dos variables utlizada en econom√≠a es la ${\bf funci√≥n de producci√≥n de Cobb-Douglas}$. Esta funci√≥n se utiliza como un modelo para representar el n√∫mero de unidades producidas al variar las cantidades de trabajo ($x$) y capital ($y$). El n√∫mero de unidades producidas est√° dado por
$$
f(x,y)=Cx^ay^{1-a}, \quad C, \, a \in \mathbb{R}, \, 0<a<1
$$

Un fabricante de juguetes estima que su funci√≥n de producci√≥n es $f(x,y)=100 x^{0.6}y^{0.4}$ donde $x$ es el n√∫mero de unidades de trabajo e $y$ es el n√∫mero de unidades de capital. Comparar el nivel de producci√≥n cuando $x=1000$ e $y=500$ con el nivel de producci√≥n $x=2000$ e $y=1000$

```python
x,y=sp.symbols('x,y')
f=100*x**(0.6)*y**(0.4)
a=1000
b=500
print(f.subs([(x,a),(y,b)]))
a=2000
b=1000
print(f.subs([(x,a),(y,b)]))
```

```python
#Curvas de nivel con Numpy
import numpy as np

import matplotlib.pyplot as plt
g=sp.lambdify((x,y),f,'numpy')
a=np.linspace(0,2000,1000)
b=np.linspace(0,2000,1000)
X,Y=np.meshgrid(a,b)
c=plt.contour(X,Y,g(X,Y))



fig,ax=plt.subplots()
cs=ax.contour(X,Y,g(X,Y))
ax.clabel(cs,inline=1,fontsize=10)
ax.set_title("con etiquetas")
```
