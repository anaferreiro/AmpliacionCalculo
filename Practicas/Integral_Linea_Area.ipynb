La base de una valla circular de radio $10\, m$ viene dada por las ecuaciones $x(t)=10\cos(t)$, $y(t)=10\sin(t)$, $0\le t\le 2\pi$. La altura de la valla viene dada por la función $f(x,y)=x^2+ 2y^2$. Se pide:

a) Representar la valla

b) Suponiendo que un litro de pintura permite cubrir $5$ metros cuadrados de valla, determinar cuántos litros de pintura se necesitan para pintar toda la valla.

c) Aproximar el área de la valla numéricamente. Calcular el error cometido

```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.art3d as art3d


ax = plt.figure().add_subplot(projection='3d')

# Prepare arrays x, y, z
t = np.linspace(0, 2*np.pi, 100)


x = 10 * np.cos(t)
y = 10 * np.sin(t)
z=0*x

ax.plot(x, y, z,lw=4,color='green')
#ax.legend()
z=x**2+2*y**2
ax.plot(x,y,z,lw=4,color='red')

"""
line=art3d.Line3D(*zip((x[0], y[0], 0), (x[0], y[0], z[0])), marker='o')#, markevery=(1, 1)) #Modo de dibujar la linea
val=zip((x[0], y[0], 0), (x[0], y[0], z[0]))
print(list(val))
recta=[ [x[0],x[0]],[y[0],y[0]],[0,z[0]] ]
print(recta)
# *recta:  unpacks a list or tuple into position arguments
line=art3d.Line3D( *recta ,marker='o' )
"""
#Codigo para pintar una linea 
#line=art3d.Line3D( xs=[x[0],x[0]], ys=[y[0],y[0]], zs=[0,z[0]] ,marker='o' )
#ax.add_line(line)

#En la proxima version de matplotlib,lo que hay implementado en siguiente bucle se puede
#sustituir por la funcion "stem"

for xi,yi,zi in zip(x,y,z):
    line=art3d.Line3D( xs=[xi,xi], ys=[yi,yi], zs=[0,zi] )
    ax.add_line(line)

plt.show()
```

```python
#Aproximación numérica de la integral

n=30
a=0; b=2*np.pi
dt=(b-a)/n
t = np.linspace(a, b, n+1)
#print(t)
tm=t+dt/2.

x = 10 * np.cos(t)
y = 10 * np.sin(t)

xm=10*np.cos(tm)
ym=10.*np.sin(tm)
zm=xm**2+2*ym**2

def c(ti):
    xi = 10 * np.cos(ti)
    yi = 10 * np.sin(ti)
    return np.array([xi,yi])

curve=[]
for ti in t:
    curve.append(c(ti))

#print(curve)
from numpy import linalg as LA


I_aprox=0.

for i in range(0, n):
    #print(i)
    segment=curve[i+1]-curve[i]    
    norma=LA.norm(segment)
    I_aprox+=zm[i]*norma
    
print("Integral aproximada= ", I_aprox,3*1000*np.pi)

```

```python
#Cálculo simbólico
import sympy as sp
t=sp.symbols('t')
c=[10*sp.cos(t),10*sp.sin(t)]
dc=[sp.diff(c[k],t) for k in [0,1]]
norm_dc=sp.sqrt(dc[0]**2+dc[1]**2)
norm_dc=sp.simplify(norm_dc)
print(norm_dc)
x,y=sp.symbols('x,y')
f=x**2+2*y**2
g=f.subs([ (x,10*sp.cos(t)),(y,10*sp.sin(t))])
Ie=sp.integrate(g*norm_dc,(t,0,2*sp.pi))
print(Ie)

print("Error= ", np.abs(Ie.evalf()-I_aprox))
```

```python
#Este código es un ejemplo 
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.art3d as art3d
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection='3d')

N = 100
theta = np.linspace(0, 2*np.pi, N, endpoint=False)
x = np.cos(theta)
y = np.sin(theta)
z = range(N)
for xi, yi, zi in zip(x, y, z):        
    line=art3d.Line3D(*zip((xi, yi, 0), (xi, yi, zi)), marker='o', markevery=(1, 1))
    ax.add_line(line)
ax.set_xlim3d(-1, 1)
ax.set_ylim3d(-1, 1)
ax.set_zlim3d(0, N)    
plt.show()
```
