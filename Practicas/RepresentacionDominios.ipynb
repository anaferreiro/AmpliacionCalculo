$\{(x,y)\in R^2/\,x^2+ y^2\ge 9\}$

```python
%matplotlib inline
import sympy as sp
x,y=sp.symbols('x,y')

from sympy import plot_implicit
plot_implicit(x**2+y**2>=9)
```

$\{(x,y)\in R^2 /\, x+y\ge 5, \, y>x^2-1 \}$


```python
x,y=sp.symbols('x,y')
fig1=plot_implicit(x+y>=5,line_color='red',show=False)
fig2=plot_implicit(y>x**2-1,line_color='green',show=False,alpha=0.1)
fig3=plot_implicit(sp.Eq(x + y, 5),line_color='white',points=1000,show=False)
fig4=plot_implicit(sp.Eq(y-x**2,-1),line_color='yellow',points=1000,show=False)
fig1.extend(fig2)
fig1.extend(fig3)
fig1.extend(fig4)
fig1.show()
#from sympy.logic.boolalg import And
#fig3=plot_implicit(And(x+y>=5,y>x**2-1),line_color='orange') #No funciona
```
